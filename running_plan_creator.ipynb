### 4ï¸âƒ£ Criar Plano de 10K com Zonas de Treino

```python
from training_zones import TrainingZones, RaceTime

# Configurar zonas baseadas em tempos de prova
zones = TrainingZones(method='jack_daniels')
zones.add_race_time("5K Recente", RaceTime.from_time_string(5.0, "22:30"))
zones.add_race_time("10K Recente", RaceTime.from_time_string(10.0, "47:15"))
zones.calculate_zones()

# Criar plano COM zonas
plano_10k = PlanGenerator.generate_plan(
    name="Treino 10K com Zonas",
    goal="10K",
    level="intermediate",
    weeks=10,
    days_per_week=4,
    training_zones=zones
)

plano_10k.set_start_date(datetime(2025, 2, 3))

print(f"âœ… Plano 10K criado com zonas personalizadas!")
print(f"ğŸ’ª VDOT: {zones.vdot:.1f}")
```

### 6ï¸âƒ£ Visualizar com GrÃ¡ficos

```python
try:
    from plot_utils import plot_weekly_volume, plot_zone_distribution_stacked, print_zone_summary
    import matplotlib.pyplot as plt
    
    # GrÃ¡fico de volume semanal
    print("ğŸ“Š Volume Semanal\n")
    fig, ax = plot_weekly_volume(plano_10k)
    plt.show()
    
    # GrÃ¡fico de distribuiÃ§Ã£o de zonas
    print("\nğŸ“Š DistribuiÃ§Ã£o de Zonas\n")
    fig, ax = plot_zone_distribution_stacked(plano_10k)
    plt.show()
    
    # Resumo textual
    print_zone_summary(plano_10k)
    
except ImportError:
    print("âš ï¸  matplotlib nÃ£o disponÃ­vel. Instale com: pip install matplotlib")
```

### 7ï¸âƒ£ AnÃ¡lise de ProgressÃ£o de Quilometragem

```python
print(f"\nğŸ“ˆ ProgressÃ£o de Quilometragem - {plano_10k.name}")
print("=" * 60)

for week in plano_10k.schedule:
    week.calculate_total_distance()
    bar_length = int(week.total_distance_km / 2)  # Escala visual
    bar = "â–ˆ" * bar_length
    print(f"Semana {week.week_number:2d}: {bar} {week.total_distance_km:.1f} km")

# EstatÃ­sticas
total_km = sum(w.total_distance_km for w in plano_10k.schedule)
media_km = total_km / len(plano_10k.schedule)
max_km = max(w.total_distance_km for w in plano_10k.schedule)

print(f"\nğŸ“Š EstatÃ­sticas:")
print(f"   Total: {total_km:.1f} km")
print(f"   MÃ©dia: {media_km:.1f} km/semana")
print(f"   Pico: {max_km:.1f} km")
```

### 8ï¸âƒ£ Criar Plano de Maratona

```python
plano_maratona = PlanGenerator.generate_plan(
    name="Minha Primeira Maratona",
    goal="Marathon",
    level="intermediate",
    weeks=16,
    days_per_week=5
)

plano_maratona.set_start_date(datetime(2025, 1, 6))

print(f"\n{'='*60}")
print(f"ğŸƒâ€â™‚ï¸ PLANO DE MARATONA")
print(f"{'='*60}")
print(f"ğŸ“‹ Nome: {plano_maratona.name}")
print(f"ğŸ“… InÃ­cio: {plano_maratona.start_date.strftime('%d/%m/%Y')}")
print(f"ğŸ Prova: {plano_maratona.get_race_date().strftime('%d/%m/%Y')}")
print(f"â±ï¸  DuraÃ§Ã£o: {plano_maratona.weeks} semanas")

# KM por fase
fase_construcao = plano_maratona.schedule[:11]  # 70% de 16 = 11 semanas
fase_manutencao = plano_maratona.schedule[11:14]
fase_taper = plano_maratona.schedule[14:]

km_construcao = sum(w.total_distance_km for w in fase_construcao)
km_manutencao = sum(w.total_distance_km for w in fase_manutencao)
km_taper = sum(w.total_distance_km for w in fase_taper)

print(f"\nğŸ“Š KM por fase:")
print(f"   ğŸ“ˆ ConstruÃ§Ã£o (semanas 1-11): {km_construcao:.1f} km")
print(f"   ğŸ’ª ManutenÃ§Ã£o (semanas 12-14): {km_manutencao:.1f} km")
print(f"   ğŸ“‰ Taper (semanas 15-16): {km_taper:.1f} km")
print(f"   ğŸ¯ TOTAL: {km_construcao + km_manutencao + km_taper:.1f} km")
```

## ğŸ’¡ Dicas para o Sucesso

### ğŸ¯ PrincÃ­pios Fundamentais

1. **ConsistÃªncia Ã© a chave**: Ã‰ melhor treinar regularmente do que fazer treinos intensos esporadicamente
2. **Escute seu corpo**: Descanse se sentir dor ou fadiga excessiva
3. **Aumente gradualmente**: Nunca aumente mais que 10% do volume semanal
4. **RecuperaÃ§Ã£o Ã© treino**: Os dias de descanso sÃ£o quando seu corpo fica mais forte
5. **NutriÃ§Ã£o importa**: Hidrate-se bem e coma adequadamente
6. **Confie no plano**: Especialmente durante o taper - resista Ã  tentaÃ§Ã£o de fazer mais

### ğŸƒ Tipos de Treino Explicados

- **ğŸŸ¢ Easy Run**: Ritmo confortÃ¡vel, consegue conversar normalmente

- **ğŸŸ  Intervals**: RepetiÃ§Ãµes rÃ¡pidas com recuperaÃ§Ã£o entre elas
- **ğŸŒˆ Fartlek**: Brincadeira com ritmo, alterna velocidades livremente
- **ğŸ”µ Long Run**: Corrida longa em ritmo fÃ¡cil para construir resistÃªncia

### âš ï¸ PrevenÃ§Ã£o de LesÃµes

- FaÃ§a aquecimento antes de treinos intensos
- Alongue apÃ³s os treinos
- Use tÃªnis adequados e troque regularmente
- Considere treinamento cruzado (nataÃ§Ã£o, ciclismo)
- FortaleÃ§a o core e membros inferiores
